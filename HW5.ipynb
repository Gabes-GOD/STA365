```python
import pymc as pm
import arviz as az
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import random
```

```python
np.random.seed(1)

theta_true = 20
tau_true = 1 / (5**2)
n = 30
data = np.random.normal(loc=theta_true, scale=1/np.sqrt(tau_true), size=n)
```

```python
normal_gamma = pm.Model()

with normal_gamma:
    # Priors
    theta = pm.Normal("theta", mu=0, sigma=100)
    tau = pm.Gamma("tau", alpha=1, beta=1)
    
    # Likelihood (observed data)
    y_obs = pm.Normal("y_obs", mu=theta, tau=tau, observed=data)
```

```python
pm.model_to_graphviz(normal_gamma)
```

```python
m = 10000
with normal_gamma:
    # draw m posterior samples
    idata = pm.sample(draws=m, chains=4, tune=1000)
```

```python
az.plot_trace(idata)
plt.tight_layout()
```

For the posterior distributions, both $\tau$ and $\theta$ have smooth unimodal posterior densities. The peak for $\tau$ is around $1/25 = 0.04$, while the peak of $\theta$ is very slightly below the true value of $20$. All of their respective chains are similar to each other.

For the time series, both $\tau$ and $\theta$ fluctuate around stable means, suggesting the sampler has converged and is mixing well.

```python
az.summary(idata, round_to=5, hdi_prob=0.95)
```

The estimated value of $\theta$ is very close to its true value of $20$, while the estimate of $\tau$ is nearly spot-on. The true value of theta and tau are well within their respective 95% hdi. 

The mcse are fairly small for both values, which means the precision of the estimate is high.

The effective sample size is around $36000$, which is fairly close to $40000$.

The code focuses on using the slice sampler to sample from the beta distribution $f$, which has support $[0,1]$. The interval $[0,1]$ is partitioned into several small intervals each of the same width $\delta$. Let $D$ be the set of all partition points. Note $D$ is a finite set. Slice sampling is an iterative algorithm. Suppose $x_0$ is chosen. In the example code, $x_0 = 0.25$. On iteration $t \geq 1$, the algorithm works as follows:

Take a sample from Unif$[0,x_{t-1}]$, call it $y_t$. Define $D_t = \{x \in D: f(x) > y_t\}$. We set
$$a = \underset{x \in D_t}{\min}(x) - \delta$$
and
$$b = \underset{x \in D_t}{\max}(x) + \delta$$
A sample from Unif$[a,b]$ is drawn, call it $x'$. If $f(x') > y_t$, then we accept $x'$ as one of our datapoints, so we'll set $x_t = x'$. 

If $f(x') \leq y$, then the interval on which we uniformly sample from will be updated. If $x' < x_{t-1}$, then set $a = x'$ and $b = b$ and redo the steps in the line above. If $x' \geq x_{t-1}$, then set $a = a$ and $b = x'$ and redo the steps in the line above. 

Once $x_t$ is obtained, the $(t)^\text{th}$ iteration terminates and the $(t+1)^\text{th}$ iteration begins.

To use slice sampling instead of MH within the Gibbs sampling strategy, we can essentially apply slice sampling twice. 

Suppose we want to sample from $p(x,y)$. Obtain $x_0$ and $y_0$ by sampling from their respective priors. On iteration $t \geq 1$, define $f_t(x) = p(x|y_{t-1})$, then apply the slice sampling algorithm described above to $f_t$. After producing a new value $x_t$, define $g_t(y) = p(y|x_t)$ and apply the slice sampling algorithm described above to $g_t$. We then get a new value $y_t$, which completes one iteration of the slicing-within-Gibbs algorithm.
